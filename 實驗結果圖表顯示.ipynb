```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = [
    (600,37.5),(1000,39.5), (1500,39.0), (2000,37.5)
]
data_2 = [
  (600,37.5),(1000,37.5), (1500,38.0), (2000,39.0)
]
data_3 = [
  (600,37.0),(1000,38.0), (1500,39.5), (2000,38.0)
]
data_4 = [
  (600,37.5),(1000,37.5), (1500,39.0), (2000,39.0)
]
def plotData(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'bert')

def plotData1(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'albert')

def plotData2(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'bert + DRCD')

def plotData3(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'albert + DRCD')

plotData(plt, data_1)
plotData1(plt, data_2)
plotData2(plt, data_3)
plotData3(plt, data_4)
plt.ylim(36, 40)
plt.title("Situation Prediction")
plt.xlabel("Samples")
plt.ylabel("Accuracy")
#plt.plot(x, data_1, label = 'class_a')
#plt.plot(x, data_2, label = 'class_b')
#plt.plot(x, data_3, label = 'class_b')
plt.legend()
```

```python
import matplotlib.pyplot as plt
x = ['bert', 'bert + DRCD', 'albert', 'albert + DRCD']
class_a = [48.16, 48.59, 48.59, 47.51]
class_b = [41.68, 42.76, 41.03, 41.68]
plt.title("Situation Prediction")
plt.xlabel("Method")
plt.ylabel("Accuracy")
plt.ylim(35, 55)
#plt.bar(x, class_a, label = 'class_a')
rects1 = plt.bar(x, class_a, label = '2000', align = "edge", color='#FFC000', width = 0.3)
rects2 = plt.bar(x, class_b, label = '1500', align = "edge", width = -0.3)
plt.legend()
for rect in rects1:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height), ha="center", va="bottom")
for rect in rects2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height), ha="center", va="bottom")
plt.show()
```
