```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = [
    (600,17.6),(800,21.3), (1000,21.8)
]
data_2 = [
  (600,20.1),(800,23.3), (1000,24.4)
]
data_3 = [
  (600,16.8),(800,24.0), (1000,26.5)
]

def plotData(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'LJ')

def plotData1(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'DRCD+LJ(512)')

def plotData2(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, label = 'DRCD+LJ(508)')

plotData(plt, data_1)
plotData1(plt, data_2)
plotData2(plt, data_3)

plt.title("Situation Prediction")
plt.xlabel("Samples")
plt.ylabel("Accuracy")
#plt.plot(x, data_1, label = 'class_a')
#plt.plot(x, data_2, label = 'class_b')
#plt.plot(x, data_3, label = 'class_b')
plt.legend()
```
